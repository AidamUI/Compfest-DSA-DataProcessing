```python
import pandas as pd

# Load the dataset
file_path = 'Klasifikasi Pengangguran Indonesia 2023.csv'
df = pd.read_csv(file_path)

# Store the original number of rows
original_rows = len(df)

# Remove rows where 'Kabupaten/Kota' is the same as 'Provinsi'
cleaned_df = df[df['Kabupaten/Kota'] != df['Provinsi']].copy()

# Store the new number of rows
new_rows = len(cleaned_df)

# Display the result
print(f"Original number of rows: {original_rows}")
print(f"Number of rows after removal: {new_rows}")
print(f"Number of rows removed: {original_rows - new_rows}")
```

```python
import pandas as pd

# Load the two datasets
df_kabupaten = pd.read_csv('Klasifikasi Pengangguran Indonesia 2023 copy.csv')
df_provinsi = pd.read_csv('Tingkat Pengangguran Terbuka Menurut Provinsi, 2025 (1).csv')

# Rename columns in the province dataset for clarity
df_provinsi.rename(columns={
    'Tingkat Pengangguran Terbuka Menurut Provinsi (Persen)': 'TPT Provinsi',
    'Persentase Angkatan Kerja Terhadap Penduduk Usia Kerja (TPAK) menurut Provinsi (Persen)': 'TPAK Provinsi'
}, inplace=True)

# Standardize province names for merging
# Create a mapping from the kabupaten file's province names to the provinsi file's names
province_map = {
    'Aceh': 'ACEH',
    'Sumatera Utara': 'SUMATERA UTARA',
    'Sumatera Barat': 'SUMATERA BARAT',
    'Riau': 'RIAU',
    'Jambi': 'JAMBI',
    'Sumatera Selatan': 'SUMATERA SELATAN',
    'Bengkulu': 'BENGKULU',
    'Lampung': 'LAMPUNG',
    'Kepulauan Bangka Belitung': 'KEP. BANGKA BELITUNG',
    'Kepulauan Riau': 'KEP. RIAU',
    'DKI Jakarta': 'DKI JAKARTA',
    'Jawa Barat': 'JAWA BARAT',
    'Jawa Tengah': 'JAWA TENGAH',
    'DI Yogyakarta': 'DI YOGYAKARTA',
    'Jawa Timur': 'JAWA TIMUR',
    'Banten': 'BANTEN',
    'Bali': 'BALI',
    'Nusa Tenggara Barat': 'NUSA TENGGARA BARAT',
    'Nusa Tenggara Timur': 'NUSA TENGGARA TIMUR',
    'Kalimantan Barat': 'KALIMANTAN BARAT',
    'Kalimantan Tengah': 'KALIMANTAN TENGAH',
    'Kalimantan Selatan': 'KALIMANTAN SELATAN',
    'Kalimantan Timur': 'KALIMANTAN TIMUR',
    'Kalimantan Utara': 'KALIMANTAN UTARA',
    'Sulawesi Utara': 'SULAWESI UTARA',
    'Sulawesi Tengah': 'SULAWESI TENGAH',
    'Sulawesi Selatan': 'SULAWESI SELATAN',
    'Sulawesi Tenggara': 'SULAWESI TENGGARA',
    'Gorontalo': 'GORONTALO',
    'Sulawesi Barat': 'SULAWESI BARAT',
    'Maluku': 'MALUKU',
    'Maluku Utara': 'MALUKU UTARA',
    'Papua Barat': 'PAPUA BARAT',
    'Papua': 'PAPUA'
}

# Apply the mapping to the kabupaten dataframe
df_kabupaten['Provinsi_merge_key'] = df_kabupaten['Provinsi'].map(province_map)

# Merge the dataframes based on the standardized province name
merged_df = pd.merge(df_kabupaten, df_provinsi, left_on='Provinsi_merge_key', right_on='Provinsi', how='left')

# Clean up by dropping the temporary merge key and the redundant province column
merged_df.drop(columns=['Provinsi_merge_key', 'Provinsi_y'], inplace=True)
merged_df.rename(columns={'Provinsi_x': 'Provinsi'}, inplace=True)


# Display the first few rows of the result
print(merged_df.head())

# Display columns to confirm the new columns are added
print("\nColumns in the new dataframe:")
print(merged_df.columns)
```

```python
# Save the merged dataframe to a new CSV file
output_filename = 'pengangguran_merged_2023.csv'
merged_df.to_csv(output_filename, index=False)

print(f"\nMerged data successfully saved to '{output_filename}'")
```
