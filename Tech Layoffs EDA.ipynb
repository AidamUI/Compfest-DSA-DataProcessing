```python
# klasifikasi kemiskinan
klasifikasi_kemiskinan = pd.read_csv('Klasifikasi Tingkat Kemiskinan di Indonesia.csv', sep=';')
klasifikasi_kemiskinan
```

```python
klasifikasi_kemiskinan.info()
```

```python
# Remove rows where all columns are NaN
klasifikasi_kemiskinan_cleaned = klasifikasi_kemiskinan.dropna(how='all')

# Display the info of the cleaned dataframe
klasifikasi_kemiskinan_cleaned.info()
```

```python
cols_to_convert = [
    'Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)',
    'Rata-rata Lama Sekolah Penduduk 15+ (Tahun)',
    'Indeks Pembangunan Manusia',
    'Umur Harapan Hidup (Tahun)',
    'Persentase rumah tangga yang memiliki akses terhadap sanitasi layak',
    'Persentase rumah tangga yang memiliki akses terhadap air minum layak',
    'Tingkat Pengangguran Terbuka',
    'Tingkat Partisipasi Angkatan Kerja'
]

# ubah koma jadi titik for decimals
for col in cols_to_convert:
    klasifikasi_kemiskinan_cleaned[col] = klasifikasi_kemiskinan_cleaned[col].str.replace(',', '.').astype(float)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = klasifikasi_kemiskinan_cleaned.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Socio-Economic Indicators')
plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# calculate average unemployment rate per province sorted
provinsi_pengangguran = klasifikasi_kemiskinan_cleaned.groupby('Provinsi')['Tingkat Pengangguran Terbuka'].mean().sort_values(ascending=False)

# seaborn bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x=provinsi_pengangguran.index, y=provinsi_pengangguran.values, palette='viridis')

plt.title('Rata-rata Tingkat Pengangguran Terbuka per Provinsi', fontsize=16)
plt.xlabel('Provinsi', fontsize=12)
plt.ylabel('Rata-rata Tingkat Pengangguran (%)', fontsize=12)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# crosstab to count occurrences of each poverty classification per province
poverty_classification_counts = pd.crosstab(klasifikasi_kemiskinan_cleaned['Provinsi'], klasifikasi_kemiskinan_cleaned['Klasifikasi Kemiskinan'])

# sort provinces by total number of kabupaten/kota
poverty_classification_counts['total'] = poverty_classification_counts.sum(axis=1)
poverty_classification_counts = poverty_classification_counts.sort_values('total', ascending=False)
poverty_classification_counts = poverty_classification_counts.drop(columns='total')


# stacked bar plot
poverty_classification_counts.plot(kind='bar', stacked=True, figsize=(15, 8), colormap='coolwarm')

plt.title('Distribusi Klasifikasi Kemiskinan per Provinsi', fontsize=16)
plt.xlabel('Provinsi', fontsize=12)
plt.ylabel('Jumlah Kabupaten/Kota', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Klasifikasi Kemiskinan', labels=['Tidak Rentan', 'Rentan'])
plt.tight_layout()
plt.show()
```

```python
klasifikasi_kemiskinan_2024 = pd.read_csv('Tingkat Pengangguran Terbuka Menurut Provinsi, 2025.csv', sep=',')
klasifikasi_kemiskinan_2024
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# sort data by unemployment rate
unemployment_sorted = klasifikasi_kemiskinan_cleaned.sort_values(by='Tingkat Pengangguran Terbuka', ascending=False)

# create figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 12))

# plot top 20 highest unemployment rates
sns.barplot(
    x='Tingkat Pengangguran Terbuka',
    y='Kab/Kota',
    data=unemployment_sorted.head(20),
    palette='Reds_r',
    ax=axes[0]
)
axes[0].set_title('Top 20 Kabupaten/Kota dengan Tingkat Pengangguran Terbuka Tertinggi', fontsize=16)
axes[0].set_xlabel('Tingkat Pengangguran Terbuka (%)', fontsize=12)
axes[0].set_ylabel('Kabupaten/Kota', fontsize=12)

# plot 20 lowest unemployment rates
sns.barplot(
    x='Tingkat Pengangguran Terbuka',
    y='Kab/Kota',
    data=unemployment_sorted.tail(20).sort_values(by='Tingkat Pengangguran Terbuka', ascending=True),
    palette='Greens_r',
    ax=axes[1]
)
axes[1].set_title('Top 20 Kabupaten/Kota dengan Tingkat Pengangguran Terbuka Terendah', fontsize=16)
axes[1].set_xlabel('Tingkat Pengangguran Terbuka (%)', fontsize=12)
axes[1].set_ylabel('Kabupaten/Kota', fontsize=12)

# adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
bali_unemployment = klasifikasi_kemiskinan_cleaned[klasifikasi_kemiskinan_cleaned['Provinsi'] == 'BALI']

# bali unemployment plot
bali_unemployment_sorted = bali_unemployment.sort_values(by='Tingkat Pengangguran Terbuka', ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(
    x='Tingkat Pengangguran Terbuka',
    y='Kab/Kota',
    data=bali_unemployment_sorted,
    palette='viridis'
)

plt.title('Tingkat Pengangguran Terbuka di Bali per Kabupaten/Kota', fontsize=16)
plt.xlabel('Tingkat Pengangguran Terbuka (%)', fontsize=12)
plt.ylabel('Kabupaten/Kota', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# clean 2024 unemployment data
# first 3 rows are headers
# last row is a summary for indonesia
unemployment_2024_cleaned = klasifikasi_kemiskinan_2024.iloc[3:-1].copy()

# select only province and february columns
unemployment_2024_cleaned = unemployment_2024_cleaned.iloc[:, :2]

unemployment_2024_cleaned.columns = ['Provinsi', 'Tingkat Pengangguran Februari']
    
unemployment_2024_cleaned['Tingkat Pengangguran Februari'] = pd.to_numeric(unemployment_2024_cleaned['Tingkat Pengangguran Februari'])

unemployment_2024_cleaned.reset_index(drop=True, inplace=True)

unemployment_2024_cleaned.head()
```

```python
# Sort data by unemployment rate
unemployment_2024_sorted = unemployment_2024_cleaned.sort_values(by='Tingkat Pengangguran Februari', ascending=False)

# bar plot using seaborn
plt.figure(figsize=(15, 8))
sns.barplot(x='Provinsi', y='Tingkat Pengangguran Februari', data=unemployment_2024_sorted, palette='viridis')

plt.title('Tingkat Pengangguran Terbuka per Provinsi (Februari 2024)', fontsize=16)
plt.xlabel('Provinsi', fontsize=12)
plt.ylabel('Tingkat Pengangguran (%)', fontsize=12)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
unemployment_2023_df = pd.read_csv('Klasifikasi Pengangguran Indonesia 2023.csv', sep=',')
unemployment_2023_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with NaN values in 'Tingkat Pengangguran Terbuka'
unemployment_2023_cleaned = unemployment_2023_df.dropna(subset=['Tingkat Pengangguran Terbuka'])

# sort data by unemployment rate
unemployment_sorted = unemployment_2023_cleaned.sort_values(by='Tingkat Pengangguran Terbuka', ascending=False)

# create figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 12))

# plot top 20 highest unemployment rates
sns.barplot(
    x='Tingkat Pengangguran Terbuka',
    y='Kabupaten/Kota',
    data=unemployment_sorted.head(20),
    palette='Reds_r',
    ax=axes[0]
)
axes[0].set_title('Top 20 Kabupaten/Kota dengan Tingkat Pengangguran Terbuka Tertinggi (2023)', fontsize=16)
axes[0].set_xlabel('Tingkat Pengangguran Terbuka (%)', fontsize=12)
axes[0].set_ylabel('Kabupaten/Kota', fontsize=12)

# plot 20 lowest unemployment rates
sns.barplot(
    x='Tingkat Pengangguran Terbuka',
    y='Kabupaten/Kota',
    data=unemployment_sorted.tail(20).sort_values(by='Tingkat Pengangguran Terbuka', ascending=True),
    palette='Greens_r',
    ax=axes[1]
)
axes[1].set_title('Top 20 Kabupaten/Kota dengan Tingkat Pengangguran Terbuka Terendah (2023)', fontsize=16)
axes[1].set_xlabel('Tingkat Pengangguran Terbuka (%)', fontsize=12)
axes[1].set_ylabel('Kabupaten/Kota', fontsize=12)

# adjust layout and show the plot
plt.tight_layout()
plt.show()
```
